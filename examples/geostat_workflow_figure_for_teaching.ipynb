Generates a data set by simulation of a random gaussian process with known parameters. This data set will then be used to test the construction and identification of a variogram model.

## Maximum de vraissemblance

Dans cette partie, on essaie d'ajuster le modèle par maximum de vraissemblance.

```python
from scipy.spatial.distance import pdist, squareform

n = len(x)
d = squareform(pdist(x))

vm = v-np.mean(v) # Ici il y a un point discutable - j'estime la moyenne en ne prenant pas de trend

# Boucle pour varier les paramètres
sill = np.linspace(1.4,2.1,50) # Les valeurs pour la boucle
L = np.zeros(sill.shape) # Initialise le vecteur de résultat à zéro

for i,s in zip( range(len(sill)), sill):

    # Calcul de la vraissemblance
    cov = geo.Covariance(s,0.8,typ='spherical')
    c = cov(d)
    c1 = np.linalg.inv(c)
    L[i] = -0.5*(n*np.log(2*np.pi) + np.log(np.linalg.det(c))
          +np.dot(v.transpose(), np.dot(c1, v))) 


plt.plot(sill,L)
plt.grid()
plt.xlabel('range')
plt.ylabel('likeliood')
plt.show()
```

```python
# Here we create a cross_validation object. The idea is to use that object to store
# the data set and then make multiple cross validation test with different models.
# So, we just initialize the object here. Internally, the function stores the required
# information and precomputes some basic statistics.

ctest = geo.CrossValid( x, v )

print( 'Variogram: ',variomod )
ctest( variomod )
```

```python
test = range(800,1900,100)
Q1_test, Q12_test, Q2_test, Cr_test, gM_test = [], [], [], [], []
for i in range(len(test)):
    vario = geo.Variogram( 0.35, test[i], typ='exponential' , nugget=0.1)
    print('--- ',vario)
    Q1,Q2,gM,Cr = ctest( vario, makeplot=False)
    Q1_test.append(Q1)
    Q12_test.append((Q2-1.0)**2)
    Q2_test.append(Q2)
    Cr_test.append(Cr)
    gM_test.append(gM)
    print(' ')
```

```python
plt.figure( figsize=(12,8) )

plt.subplot(221)
plt.plot(test,Q1_test,'.-')
plt.title('Q1')

plt.subplot(222)
plt.plot(test,Q12_test,'.-')
plt.title('Q2')

plt.subplot(223)
plt.plot(test,gM_test,'.-')
plt.title('gM')

plt.subplot(224)
plt.plot(test,Cr_test,'.-')
plt.title('Cr')

plt.show()
```
