# Unconditional stochastic simulations using Cholesky decomposition

The aim of this document is to test the generation of stochastic realizations of a Gaussian process using Cholesky decomposition as coded in the simulation package.

```python
import numpy as np
import matplotlib.pyplot as plt
import geostat as geo
```

##  Defines the geometry of the problem

For this example, the geometry that is considered is a one dimensional regular and cartesian grid. The number of nodes $n$ is given.

```python
n = 139*2  # Number of nodes

x = np.linspace(0, 200, n)  # Creates n regular coordinates from 0 to 200
X = np.zeros((n,2))         # Assembles x values as a 2 D coordinate matrix 
X[:,0] = x
```

## Generates n simulations

Here we use the simulation function defined outside of this worksheet.

If one uses a gaussian covariance and a range that is too large, then the covariance matrix may not be exactly definite positive and the Cholesky decomposition may not work.


```python
covar = geo.Covariance(rang=10,sill=1,typ='spherical')

y = simulation.unconditionnal_lu( X, covar,   nsimuls=3 )
```
