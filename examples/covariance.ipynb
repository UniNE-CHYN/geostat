# Shows how to compute covariance using the geostat class



```python
import numpy as np
import matplotlib.pyplot as plt
import geostat as geo
```

```python
for t in ["spherical"]: #, "gaussian"]:
    
    h = np.linspace(0,3,100)
    
    covar = geo.Covariance(sill=2.0,rang=2,typ=t)
    gamma = geo.Variogram(sill=2.0,rang=2,typ=t)
    c = covar(h)
    g = gamma(h)

    plt.plot(h,c,h,g)
    plt.grid()
    plt.xlabel('h')
    plt.ylabel('C(h)')
    plt.title(t)
    plt.savefig('tmp.png', bbox_inches='tight')
    plt.show()
```
