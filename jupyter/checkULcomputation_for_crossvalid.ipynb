```python
import numpy as np
from scipy.special import gamma

def fch(Q2,n):
    k = n-1
    f = np.power(k, k/2) * np.power(Q2, (k-2)/2) * np.exp( -k*Q2/2 )
    f = f/( np.power(2,k/2) * gamma(k/2))
    return f
    
Q2 = np.arange( 0, 2.5, 0.01)    
p = fch(Q2,200) 
c = np.cumsum(p)*0.01

import matplotlib.pyplot as plt
plt.plot( Q2, c)
plt.show()
```

```python
from scipy.stats import chi2

def UL(n):
    df = n-1
    L = chi2.ppf( 0.025, df) / df 
    U = chi2.ppf( 0.975, df) / df
    return L, U
    

n = 101
print( UL(n))
```
